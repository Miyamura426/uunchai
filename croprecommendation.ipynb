```python
# Define quantile thresholds for each column
low, high = df["P"].quantile([0.01, 0.9])
low1, high1 = df["K"].quantile([0.01, 0.9])
low2, high2 = df["ph"].quantile([0.1, 0.9])
low3, high3 = df["temperature"].quantile([0.1, 0.9])
low4, high4 = df["rainfall"].quantile([0.1, 0.9])

# Create masks for each column
mask_P=df["P"].between(low,high)
mask_K = df["K"].between(low1, high1)
mask_ph = df["ph"].between(low2, high2)
mask_temperature = df["temperature"].between(low3, high3)
mask_rainfall = df["rainfall"].between(low4, high4)

# Combine all masks
combined_mask = mask_K & mask_ph & mask_temperature & mask_rainfall

# Filter the DataFrame
df_filtered = df[combined_mask]
```

```python
df=df[mask_K & mask_ph & mask_temperature & mask_rainfall]
```

```python
low5, high5 = df["humidity"].quantile([0.1, 0.99])
mask_humidity = df["humidity"].between(low5, high5)
df=df[mask_humidity]
```
